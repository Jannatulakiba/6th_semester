```python
import random

# Function to estimate the value of pi using Monte Carlo methods
def estimate_pi(num_samples):
    num_inside = 0
    for i in range(num_samples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            num_inside += 1
    pi_estimate = 4 * num_inside / num_samples
    return pi_estimate

# Estimate the value of pi using 10,000 samples
pi_estimate = estimate_pi(10000)
print("Estimated value of pi:", pi_estimate)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random points and plot them
num_samples = 10000
x = np.random.uniform(-1, 1, num_samples)
y = np.random.uniform(-1, 1, num_samples)
plt.plot(x, y, 'k.', markersize=1)

# Plot the quarter circle of radius 1
t = np.linspace(0, np.pi/2, 100)
xc = np.cos(t)
yc = np.sin(t)
plt.plot(xc, yc, 'r-', linewidth=2)

# Set the axis limits and labels
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel('x')
plt.ylabel('y')

# Show the plot and the estimated value of pi
plt.title('Estimate of Pi: {:.4f}'.format(pi_estimate))
plt.show()
```
