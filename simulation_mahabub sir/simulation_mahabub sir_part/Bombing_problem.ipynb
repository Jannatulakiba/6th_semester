```python
import random
import matplotlib.pyplot as plt

hit_count = 0
miss_count = 0

N = 500   # total simulations

convergence = []   # store strike % over time

for i in range(1, N + 1):
    x_random = random.uniform(-1.5, 1.5)
    x_val = 500 * x_random

    y_random = random.uniform(-1.5, 1.5)
    y_val = 300 * y_random

    if (abs(x_val) <= 500) and (abs(y_val) <= 300):
        hit_count += 1
    else:
        miss_count += 1

    # Compute strike % at each step
    strike_percentage = hit_count / i * 100
    convergence.append(strike_percentage)

# Final result
print("Final Strike Percentage:", convergence[-1])

# Plot convergence graph
plt.figure(figsize=(8,5))
plt.plot(convergence)

plt.axhline(convergence[-1])  # final stabilized value

plt.xlabel("Number of Trials")
plt.ylabel("Strike Percentage (%)")
plt.title("Real-Time Convergence of Strike Percentage")

plt.show()
```
