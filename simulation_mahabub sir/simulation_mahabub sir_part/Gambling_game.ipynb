```python
import random

N = 10000
total_money = 0
wins = 0

for _ in range(N):
    dice = random.randint(1, 6)

    if dice == 6:
        total_money += 10
        wins += 1
    else:
        total_money -= 2

# Results
win_probability = wins / N
average_profit = total_money / N

print(f"Win Probability: {win_probability:.4f}")
print(f"Average Profit per Game: {average_profit:.4f}")
```

```python
import random
import matplotlib.pyplot as plt

# Number of simulations
N = 500

profits = []
cumulative_avg = []

total_profit = 0

for i in range(1, N + 1):
    diff = 0
    flips = 0

    # Play one game
    while abs(diff) < 3:
        toss = random.choice(["H", "T"])
        flips += 1

        if toss == "H":
            diff += 1
        else:
            diff -= 1

    # Profit calculation
    profit = 8 - flips
    profits.append(profit)

    total_profit += profit
    cumulative_avg.append(total_profit / i)

# Final decision
print(f"Average Profit after {N} simulations: {cumulative_avg[-1]:.4f}")

# Plot
plt.figure(figsize=(8,5))
plt.plot(cumulative_avg)

plt.axhline(0)  # break-even line

plt.xlabel("Number of Simulations")
plt.ylabel("Average Profit")
plt.title("Monte Carlo Simulation of Coin Gambling Game")

plt.show()
```
